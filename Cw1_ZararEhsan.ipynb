**Part A**

Yield curves are graphs with yield against maturity, represented as a percentage against time graph. For this coursework the financial instruments I will be taking for consideration is swaps and bonds, at different maturities. The yield curve is dependent on different maturities, as a singular maturity gives you just one yield, which by itself cannot possibly be used to make a curve, therefore multiple maturities are required to make a yield curve. There is also the question of why different maturities exist in the first place, it is because their pricing is affected by different factors such as interest rate risk, inflation risk, default risk etc. Bonds are in simple terms, a form of a loan but it has certain differences, such as being able to be traded and fixed coupon terms. A bond is when a lender gives money to an entity (government or corporate), in exchange for a promise for coupons, which are interest rate payments at fixed terms, and the principal back at the end of the maturity. There are different types of bonds too such as zero-coupon bonds, zero-coupon bonds as the name suggests do not pay coupons, rather the bond is bought at a discounted rate i.e. the price and a singular face value payment at maturity, with a specific formula to calculate the yield. Since the yield from a zero-coupon bond is calculated from this single payment at the end, it makes the yield calculation simpler to perform.

A clarification that needs to be made is that even though the coupon rates are fixed, bonds can be traded and therefore priced differently at different points, hence why yields can change.



---



The second instrument for this yield curve is an interest rate swap, which is an over the counter agreement between two parties. These agreements are tailored to meet specific needs of the parties involved, allowing them to hedge against interest rates. A swap happens when entity A and entity B want to both get involved, however entity A wants to hedge on a fixed rate and entity B wants to hedge on a floating rate. A notional is then agreed upon which is then used to estimate a future projection of the cashflow, the initial payment is based on the current floating rate (floating leg), the subsequent payments are then calculated from the forward rate implied by discounting factors. The discount factors are also applied on the projected cashflow to workout the present value of the payment. A formula is then used to calculate the swap rate, which is the yield, which also is by definition the rate that balances the fixed leg and the floating leg of the swap. This swap rate being the projected rate that is supposed to keep things fair between the two parties involved in the swap is then used as the fixed rate for the fixed leg. Obviously, in real life you will never have a perfect projection, therefore the swap rate is the tell you can use to determine which party benefits more from the swap ie if the party offering the floating rate payment ended up paying less than the swap rate, they were in benefit and vice versa.

Yield curve fitting is the practice of producing a continuous yield curve using discrete data. The traditional method to doing this is either a fit of a mathematical model function or a splines interpolation, the paper argues the advantage of Artificial intelligence algorithms being its ability to avoid any assumptions for the mathematical model functions. The first algorithm used is a three-layer feed forward neural network with a non-linear sigmoid activation function, they identified the hyperparameter to be the number of neurons in the hidden layer. After which they correctly identified the basis of any machine learning or AI technique, which is the objective for the model to be able to generalise the data i.e. not overfit and not underfit. The second AI technique they mention is a Support Vector Machine, which they used a wavelet kernel and the Gaussian radial basis kernel. They also identified a hyperparameter for SVMs which is the kernel width. The goodness of fit and the interpolation is then tested in sections 6 and 7 respectively, in section 6 they state that splines has the lowest RMSE, however they argue that this doesnt indicate much because splines is a piecewise fit using the data as knot points and consequently leading to a low RMSE. In section 7 they state that neural networks are surprisingly stable when using the leave one out method to interpolate, noting that when reducing the number of points to 3, it still yields a relatively good result.

Long short-term memory is a deep learning method, more specifically a Recurrent neural network. With regards to data the reason why PCA may improve the prediction accuracy is because PCA is a method to reduce the dimensionality of a large dataset, however it is important to note how this is done and it is not simply a random deletion of input variables of the data, PCA uses calculations to workout which variables hold the most weight in terms of affecting the data and then gets rid of the noise. Specific to this study data was selected from 80 stocks, they used approximately 4 hidden layers with around a thousand nodes in each of them and using a sigmoid function in each of them. The study states it took around 600 iterations of training for PCA-LSTM to reach convergence, which even if it was the fastest among the others it still might have taken a while, the time to produce results would have been fast once the model had been trained and the results show that among PCA-LSTM, LSTM and back propagation neural network, PCA-LSTM had the smallest error. **This section is to format the datasets so that each dataset has the same date format and is indexed using this date column and also to label the remaining column so that it can be differentiated based on its maturity**

```python
def clean_df(df,column_type):
  maturities = ['2','7','10','30']
  for i, df in enumerate(df):
    if 'observation_date' in df.columns:
      df['Date'] = pd.to_datetime(df['observation_date'])
      df.drop(columns=['observation_date'], inplace=True)
    elif 'Date' in df.columns:
      df['Date'] = pd.to_datetime(df['Date'])

    df.set_index('Date', inplace=True)
    df.sort_index(ascending=True, inplace=True)

    maturity_name=maturities[i]

    column_name=f"{column_type}_{maturity_name}"
    df.columns=[column_name]
    cleaned_df.append(df)
```

**This section prepares the final dataset, by splitting it for training and testing in chronological order to avoid look ahead bias**

```python
data_values = final_dataset.values
```

```python
train_size = int(len(data_values) * 0.7)
test_size = len(data_values) - train_size
train, test = data_values[0:train_size,:], data_values[train_size:len(data_values),:]
print(len(train), len(test))
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

**This section handles the lookback window to fit the LSTM models needs**

```python
look_back=75
def create_df(df,look_back):
  X,Y = [],[]
  for i in range(look_back, len(df)):
    X.append(df[i-look_back:i,:])
    Y.append(df[i,:])

  return np.array(X), np.array(Y)
```

```python
x_train, y_train = create_df(train_scaled,look_back)
x_test, y_test = create_df(test_scaled,look_back)
```

**This section is for configuring, fitting and training the model on the data**

```python
model = Sequential()
model.add(LSTM(64, input_shape=(look_back,8),return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(8))
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
training=model.fit(x_train, y_train, epochs=50, batch_size=32,validation_data=(x_test,y_test), verbose=2)
```

**This section is to get the predictions from the model and test its error using RMSE**

```python
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
```

```python
train_predict = scaler.inverse_transform(train_predict)
train_y = scaler.inverse_transform(y_train)
test_predict = scaler.inverse_transform(test_predict)
test_y = scaler.inverse_transform(y_test)
```

```python
trainScore = np.sqrt(mean_squared_error(train_y, train_predict))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(test_y, test_predict))
print('Test Score: %.2f RMSE' % (testScore))
```

**This section is to make a prediction for the specified number of datas ahead, for each maturity and for each financial instrument to make a final weighted yield curve**

Do note, the further ahead you try to predict the less accurate the results become due to the recursive nature of the results

```python
days_to_predict=4 # Change this value in accordance with how many days ahead you want to predict
```

```python
full_data_scaled = scaler.transform(data_values)
current_batch = full_data_scaled[-75:].reshape((1, 75, 8))
predictions_scaled = []

# 2. Run the Recursive Loop
print(f"Simulating {days_to_predict} days into the future...")

for i in range(days_to_predict):

    current_pred = model.predict(current_batch, verbose=0)

    predictions_scaled.append(current_pred[0])


    new_feed = np.append(current_batch[:, 1:, :], [current_pred], axis=1)
    current_batch = new_feed

predictions_actual = scaler.inverse_transform(np.array(predictions_scaled))
```

```python
final_pred = predictions_actual[-1]
bond_pred_final = final_pred[0:4]
swap_pred_final = final_pred[4:8]

tenors = ['2Y', '7Y', '10Y', '30Y']
```

```python
latest_date = final_dataset.index.max()
print(f"The latest date in the database is: {latest_date}")
```

```python
bond_weight=0.6
swap_weight=0.4

weighted_yield=(bond_weight*bond_pred_final)+(swap_weight*swap_pred_final)


target_day = latest_date + pd.Timedelta(days=days_to_predict)

plt.figure(figsize=(10, 6))

plt.plot(tenors, weighted_yield, label='Weighted Yield', marker='o', linewidth=2)
plt.plot(tenors, bond_pred_final, label='Bond Prediction', linestyle='--', marker='x')
plt.plot(tenors, swap_pred_final, label='Swap Prediction', linestyle='-.', marker='s')

plt.grid(True, alpha=0.3)

plt.xlabel('Tenors')
plt.ylabel('Yield')
plt.title(f'Yield Curve Predictions ({target_day.strftime("%Y-%m-%d")})')
plt.legend()
plt.show()
```
